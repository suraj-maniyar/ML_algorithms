```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
k = 3
num = 100

dimension = 2

class1 = [50,50]
class2 = [120,50]
class3 = [50, 120]

cov = [[150,0],[0,150]]

x1 = np.random.multivariate_normal(class1, cov, num)
x2 = np.random.multivariate_normal(class2, cov, num)
x3 = np.random.multivariate_normal(class3, cov, num)

x = np.concatenate([x1, x2, x3], axis=0)

X = x[:,0]
Y = x[:,1]
colors = cm.rainbow(np.linspace(0, 1, k))

centers = 180.0*np.random.rand(k,dimension)

cost_arr = []

print(x.shape)
plt.plot(x[:,0], x[:,1], 'o')
plt.plot(centers[:, 0], centers[:, 1], 'v', markersize=10)
```

```python
class kmeans:
    def __init__(self, k, dim):
        self.k = k
        self.dim = dim
        self.clusters = []
        self.centers = 180.0*np.random.rand(self.k, self.dim)
        self.cost = 0
        self.colors = cm.rainbow(np.linspace(0, 1, self.k))
        
    def distanceof(self, a, b):
        return np.linalg.norm(a-b)

    def l2(self, a,b):
        return np.sqrt(np.sum( np.square(a-b) ))
    
    def update(self, x): 
        # get the distance from each point the cluster centers 
        distance = np.zeros((x.shape[0], self.k))
        for i in range(x.shape[0]):
            for k_index in range(self.k):
                distance[i, k_index] = self.distanceof(x[i], self.centers[k_index])
    
        # Assign cluster to each point by taking the nearest cluster center
        self.clusters = np.zeros(x.shape[0])
        for i in range(len(distance)):
            self.clusters[i] = np.argmax(distance[i])
        
        # Update the cluster centers by taking the mean of the point belonging to its own cluster
        self.centers = np.zeros((self.k, self.dim))
        for class_index in range(self.k):
            nums = 0
            for i in range(x.shape[0]):
                if(self.clusters[i] == class_index):
                    self.centers[class_index] += x[i]
                    nums += 1
            self.centers[class_index] /= nums
        
        # Calculate the cost by taking intra cluster distance (Should be minimum)
        self.cost = 0
        for i in range(x.shape[0]):
            for k_index in range(self.k):
                if(self.clusters[i] == k_index):
                    self.cost += self.l2(self.centers[k_index], x[i])

    # Run the iteration multiple times                 
    def fit(self, x, n_iter):
        cost_arr = []
        for i in range(n_iter):
            self.update(x)
            cost_arr.append(self.cost)
        return cost_arr
    
    # Display clusters with different colors
    def display(self, x):
        for k_index in range(self.k):
            X = [ x[i,0] for i in range(x.shape[0]) if self.clusters[i] == k_index ]
            Y = [ x[i,1] for i in range(x.shape[0]) if self.clusters[i] == k_index ]
            plt.scatter(X, Y, color = self.colors[k_index])
        plt.plot(self.centers[:, 0], self.centers[:, 1], 'v', markersize=15)
        print(self.centers)
```

```python
km = kmeans(k=3, dim=2)
```

```python
print('cost = ', km.fit(x, n_iter=1)[0])
km.display(x)
```
